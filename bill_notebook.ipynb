```python
from bs4 import BeautifulSoup
import requests
import os
import numpy as np
import pandas as pd
```

```python
url = 'https://txschools.gov/data/schools.json?v=.0004'
headers = {'User-Agent': 'l'}
response = requests.get(url, headers=headers)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup.prettify
```

```python
article = soup.find_all('a', class_='table') #(tag beginning, class = 'class name')
article
```

```python
lists = list(range(1, len(article) - 1))
```

```python
hs = []
for i in lists:
    test = article[i].find('a')
    hs.append(test)
```

```python
len(hs)
```
